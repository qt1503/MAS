```python
import json
from langsmith import Client
import pandas as pd
```

```python
client = Client()
```

```python
experiment_name = "" 

runs = list(client.list_runs(project_name=experiment_name, execution_order=1))

data = []
count = 0
for run in runs:
    is_correct = None
    feedback_list = client.list_feedback(run_ids=[run.id])
    for fb in feedback_list:
        if fb.key == "is_correct":
            is_correct = fb.score
    count+=1
    row = {
        "run_id": run.id,
        "error": run.error,
        "latency_sec": (run.end_time - run.start_time).total_seconds() if run.end_time and run.start_time else None,
        "total_cost": run.total_cost,
        "input_tokens": run.prompt_tokens,
        "output_tokens": run.completion_tokens,
        "total_tokens": run.total_tokens,
        "is_correct": is_correct,
    }
    data.append(row)

df_gsm8k = pd.DataFrame(data)
```

```python
def clean_dataframe_for_json(df):
    """
    L√†m s·∫°ch dataframe ƒë·ªÉ xu·∫•t ra JSON b·∫±ng c√°ch x·ª≠ l√Ω c√°c l·ªói m√£ h√≥a UTF-8
    
    Tham s·ªë:
        df: pandas DataFrame c·∫ßn l√†m s·∫°ch
        
    Tr·∫£ v·ªÅ:
        pandas DataFrame v·ªõi d·ªØ li·ªáu ƒë√£ ƒë∆∞·ª£c l√†m s·∫°ch
    """
    df_clean = df.copy()
    
    # Chuy·ªÉn ƒë·ªïi c√°c c·ªôt c√≥ v·∫•n ƒë·ªÅ th√†nh string v√† x·ª≠ l√Ω l·ªói encoding
    for col in df_clean.columns:
        if df_clean[col].dtype == 'object':
            df_clean[col] = df_clean[col].astype(str)
            # Thay th·∫ø c√°c k√Ω t·ª± c√≥ v·∫•n ƒë·ªÅ
            df_clean[col] = df_clean[col].str.encode('utf-8', errors='ignore').str.decode('utf-8')
    
    return df_clean
```

```python
def save_dataframes_to_json(df_tatqa, df_gsm8k, df_tabmwp, method):
    """
    L∆∞u c√°c dataframe ƒë√£ ƒë∆∞·ª£c l√†m s·∫°ch th√†nh file JSON v√†o folder results
    
    Tham s·ªë:
        df_tatqa: DataFrame d·ªØ li·ªáu TATQA
        df_gsm8k: DataFrame d·ªØ li·ªáu GSM8K  
        df_tabmwp: DataFrame d·ªØ li·ªáu TABMWP
    """
    # L√†m s·∫°ch c√°c dataframe tr∆∞·ªõc khi l∆∞u
    df_tatqa_clean = clean_dataframe_for_json(df_tatqa)
    df_gsm8k_clean = clean_dataframe_for_json(df_gsm8k)
    df_tabmwp_clean = clean_dataframe_for_json(df_tabmwp)

    # L∆∞u t·ª´ng file JSON v·ªõi error handling
    try:
        df_tatqa_clean.to_json(f"results/{method}_TATQA.json", orient="records", indent=2, force_ascii=False)
        print("‚úÖ TATQA JSON ƒë√£ l∆∞u th√†nh c√¥ng")
    except Exception as e:
        print(f"‚ùå L·ªói khi l∆∞u TATQA JSON: {e}")

    try:
        df_gsm8k_clean.to_json(f"results/{method}_GSM8K.json", orient="records", indent=2, force_ascii=False)
        print("‚úÖ GSM8K JSON ƒë√£ l∆∞u th√†nh c√¥ng")
    except Exception as e:
        print(f"‚ùå L·ªói khi l∆∞u GSM8K JSON: {e}")

    try:
        df_tabmwp_clean.to_json(f"results/{method}_TABMWP.json", orient="records", indent=2, force_ascii=False)
        print("‚úÖ TABMWP JSON ƒë√£ l∆∞u th√†nh c√¥ng")
    except Exception as e:
        print(f"‚ùå L·ªói khi l∆∞u TABMWP JSON: {e}")
    
    print("\nüéâ Ho√†n th√†nh vi·ªác l∆∞u t·∫•t c·∫£ file JSON!")
```

```python
save_dataframes_to_json(df_tatqa, df_gsm8k, df_tabmwp, "")
```
